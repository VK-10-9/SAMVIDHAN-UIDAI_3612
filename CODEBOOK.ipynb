## Core API Endpoints

### National Overview
Get high-level national statistics

````python
# Example: Fetch national overview
def get_national_overview():
    endpoint = f"{BASE_URL}/api/national-overview"
    response = requests.get(endpoint)
    
    if response.status_code == 200:
        data = response.json()
        print("ğŸ“Š National Overview:")
        for key, value in data.items():
            print(f"  {key}: {value}")
        return data
    else:
        print(f"âŒ Error: {response.status_code}")
        return None

# Uncomment to test
# overview = get_national_overview()
````

### State Analytics
Get state-specific data and comparisons

````python
# Example: Fetch state data
def get_state_data(state_code):
    endpoint = f"{BASE_URL}/api/analytics/states/{state_code}"
    response = requests.get(endpoint)
    
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print(f"âŒ Error fetching state {state_code}: {response.status_code}")
        return None

# Example usage
# delhi_data = get_state_data('DL')
# maharashtra_data = get_state_data('MH')
````

### Enrollment Timeline
Get historical enrollment trends

````python
# Example: Fetch enrollment trends
def get_enrollment_timeline(days=30):
    endpoint = f"{BASE_URL}/api/enrollment-timeline"
    params = {'days': days}
    response = requests.get(endpoint, params=params)
    
    if response.status_code == 200:
        data = response.json()
        df = pd.DataFrame(data)
        return df
    else:
        print(f"âŒ Error: {response.status_code}")
        return None

# Example usage
# timeline = get_enrollment_timeline(days=90)
````

### Anomaly Detection
Get detected anomalies in the data

````python
# Example: Fetch anomalies
def get_anomalies(severity='all', limit=100):
    endpoint = f"{BASE_URL}/api/anomalies/list"
    params = {'severity': severity, 'limit': limit}
    response = requests.get(endpoint, params=params)
    
    if response.status_code == 200:
        data = response.json()
        df = pd.DataFrame(data)
        print(f"ğŸ” Found {len(df)} anomalies")
        return df
    else:
        print(f"âŒ Error: {response.status_code}")
        return None

# Example usage
# anomalies_high = get_anomalies(severity='high')
````

# Intelligence Frameworks

## ADIF - Aadhaar Data Integrity Framework
Ensures data quality and consistency

````python
# ADIF: Data Integrity Checks
def check_duplicate_records(limit=50):
    endpoint = f"{BASE_URL}/api/signals/duplicates"
    response = requests.get(endpoint, params={'limit': limit})
    
    if response.status_code == 200:
        data = response.json()
        print(f"ğŸ” ADIF - Duplicate Detection: {len(data)} records found")
        return data
    else:
        print(f"âŒ Error: {response.status_code}")
        return None

# Example usage
# duplicates = check_duplicate_records()
````

## IRF - Identity Resilience Framework
Biometric verification and aging assessment

````python
# IRF: Multi-factor verification
def verify_identity_multi_factor(record_id):
    endpoint = f"{BASE_URL}/api/irf/multi-factor"
    params = {'record_id': record_id}
    response = requests.get(endpoint, params=params)
    
    if response.status_code == 200:
        data = response.json()
        print(f"âœ… IRF - Verification Status:")
        for key, value in data.items():
            print(f"  {key}: {value}")
        return data
    else:
        print(f"âŒ Error: {response.status_code}")
        return None

# Example usage
# verify_identity_multi_factor('AADHAAR_ID_123456')
````

## AFIF - Aadhaar Forensic Intelligence Framework
Fraud detection and network analysis

````python
# AFIF: Hub/network detection
def analyze_fraud_hubs(threshold=5):
    endpoint = f"{BASE_URL}/api/afif/hub-analysis"
    params = {'threshold': threshold}
    response = requests.get(endpoint, params=params)
    
    if response.status_code == 200:
        data = response.json()
        df = pd.DataFrame(data)
        print(f"ğŸš¨ AFIF - Fraud Hubs Detected: {len(df)}")
        return df
    else:
        print(f"âŒ Error: {response.status_code}")
        return None

# Example usage
# hubs = analyze_fraud_hubs(threshold=3)
````

## PROF - Public Resource Optimization Framework
Migration analysis and demand forecasting

````python
# PROF: Migration Pressure Index
def get_migration_pressure_index(state_code):
    endpoint = f"{BASE_URL}/api/prof/mpi"
    params = {'state': state_code}
    response = requests.get(endpoint, params=params)
    
    if response.status_code == 200:
        data = response.json()
        print(f"ğŸ“Š PROF - Migration Pressure Index for {state_code}:")
        print(f"  MPI Score: {data.get('mpi_score')}")
        print(f"  Risk Level: {data.get('risk_level')}")
        return data
    else:
        print(f"âŒ Error: {response.status_code}")
        return None

# Example usage
# mpi = get_migration_pressure_index('UP')
````

## AMF - Aadhaar Mobility Framework
Population mobility tracking and analysis

````python
# AMF: Mobility classification
def classify_mobility_tier(record_id):
    endpoint = f"{BASE_URL}/api/amf/mobility-tier"
    params = {'record_id': record_id}
    response = requests.get(endpoint, params=params)
    
    if response.status_code == 200:
        data = response.json()
        print(f"ğŸš— AMF - Mobility Classification:")
        for key, value in data.items():
            print(f"  {key}: {value}")
        return data
    else:
        print(f"âŒ Error: {response.status_code}")
        return None

# Example usage
# mobility = classify_mobility_tier('AADHAAR_ID_789012')
````

## PPAF - Privacy-Preserving Analytics Framework
Differential privacy and secure data analysis

````python
# PPAF: Differential Privacy query
def query_with_differential_privacy(query, epsilon=1.0):
    endpoint = f"{BASE_URL}/api/ppaf/differential-privacy"
    payload = {
        'query': query,
        'epsilon': epsilon
    }
    response = requests.post(endpoint, json=payload)
    
    if response.status_code == 200:
        data = response.json()
        print(f"ğŸ” PPAF - Privacy-Preserving Query Result:")
        print(f"  Result: {data.get('result')}")
        print(f"  Noise Added: {data.get('noise_level')}")
        return data
    else:
        print(f"âŒ Error: {response.status_code}")
        return None

# Example usage
# result = query_with_differential_privacy('SELECT COUNT(*) FROM enrollments', epsilon=1.0)
````

# Advanced Usage

## Batch API Operations

````python
# Batch fetch data for multiple states
def get_all_states_data(state_codes):
    results = {}
    for state in state_codes:
        try:
            endpoint = f"{BASE_URL}/api/analytics/states/{state}"
            response = requests.get(endpoint)
            if response.status_code == 200:
                results[state] = response.json()
                print(f"âœ… Fetched data for {state}")
        except Exception as e:
            print(f"âŒ Error for {state}: {str(e)}")
    return results

# Example usage
# states = ['DL', 'MH', 'TN', 'KA', 'UP', 'WB']
# all_states_data = get_all_states_data(states)
````

## Data Analysis and Visualization

````python
# Example: Prepare data for analysis
def prepare_analysis_dataset(state_codes):
    data_frames = []
    
    for state in state_codes:
        try:
            endpoint = f"{BASE_URL}/api/analytics/states/{state}"
            response = requests.get(endpoint)
            if response.status_code == 200:
                df = pd.DataFrame(response.json())
                df['state'] = state
                data_frames.append(df)
        except Exception as e:
            print(f"Error: {str(e)}")
    
    if data_frames:
        combined_df = pd.concat(data_frames, ignore_index=True)
        return combined_df
    return None

# Example usage
# analysis_df = prepare_analysis_dataset(['DL', 'MH', 'TN'])
# print(analysis_df.describe())
````

# Troubleshooting

## Common Issues & Solutions

### Issue 1: Connection Refused
```
Error: Connection refused to localhost:8000
```
**Solution:** Make sure backend server is running:
```bash
cd backend
python main.py
```

### Issue 2: CORS Errors
```
Error: CORS policy blocked request
```
**Solution:** CORS is configured in FastAPI backend. If you see this, ensure:
- Backend is running on port 8000
- Frontend is running on port 3000

### Issue 3: Timeout Errors
**Solution:** Large queries may timeout. Try:
- Reducing the date range
- Using pagination with `limit` parameter
- Checking server logs for specific errors

### Issue 4: Data Not Loading
**Solution:**
1. Check if CSV files exist in `dataset/clean/`
2. Verify file permissions
3. Check backend logs for parsing errors
